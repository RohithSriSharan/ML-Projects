```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("ml_data_sets/loan/train.csv")
df_test = pd.read_csv("ml_data_sets/loan/test.csv")
```

```python
df_train.head()
```

### Training Data Pre Processing

```python
df_train.isnull().sum()
```

```python
df_train_num = df_train.select_dtypes(include='float').columns.tolist()
print(df_train_num)
df_train_cat = df_train.select_dtypes(include='object').columns.tolist()
print(df_train_cat)
```

```python
df_train_num = ['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
df_train_cat = [ 'Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']

df_train[df_train_num] = df_train[df_train_num].fillna(df_train[df_train_num].mean())
df_train.isnull().sum()
```

```python
for col in df_train_cat:
    df_train[col] = df_train[col].fillna(df_train[col].mode()[0])
```

```python
df_train.isnull().sum()
```

```python
# Encoding

from sklearn.preprocessing import OneHotEncoder
y = df_train['Loan_Status']
df_train = df_train.drop(columns=['Loan_Status', 'Loan_ID'])

encoder = OneHotEncoder(drop='first', handle_unknown = 'ignore', sparse_output=False)

encoded = encoder.fit_transform(df_train[df_train_cat])

encoded_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out(df_train_cat), index = df_train.index)

encoded_features = pd.concat([df_train.drop(columns=df_train_cat), encoded_df], axis= 1  )
```

```python
from sklearn.preprocessing import StandardScaler
scale = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
scaler = StandardScaler()

scaled_values = scaler.fit_transform(encoded_features[scale])

encoded_features[scale] = scaled_values 
```

```python
encoded_features.head()
```

### Test Data Pre Processing

```python
df_test.isnull().sum()
```

```python
df_test = df_test.drop(columns='Loan_ID')
df_test_float = df_test.select_dtypes(include='float').columns.tolist()

df_test_int = df_test.select_dtypes(include='int').columns.tolist()
df_test_cat = df_test.select_dtypes(include='object').columns.tolist()
df_num = df_test_float + df_test_int
df_test[df_test_num] = df_test[df_test_num].fillna(df_test[df_test_num].mean())
for col in df_test_cat:
    df_test[col] = df_test[col].fillna(df_test[col].mode()[0])
df_test.isnull().sum()
```

```python


encoded_test = encoder.transform(df_test[df_test_cat])

df_test_encoded = pd.DataFrame(encoded_test, columns = encoder.get_feature_names_out(df_test_cat), index = df_test.index )

encoded_test_features = pd.concat([df_test.drop(columns = df_test_cat), df_test_encoded ], axis = 1)

```

```python
encoded_test_features.head()
```

```python

test_scaled = scaler.transform(encoded_test_features[scale])

encoded_test_features[scale] = test_scaled

encoded_test_features.head()
```

```python

y = y.map({
    'Y':1,
    'N':0
})
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(
    encoded_features, y, test_size=0.2, random_state=42, stratify=y
)


model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

```

```python
y_pred = model.predict(X_test)
```

```python
test_predictions = model.predict(encoded_test_features)
```

```python
print("Train-test split predictions (y_pred):")
print(y.value_counts()) 
print(pd.Series(y_pred).value_counts())

print("Final test predictions:")
print(pd.Series(test_predictions).value_counts())
```

```python
from sklearn.metrics import classification_report

cr= classification_report(y_test,y_pred )
print(cr)
```

```python
model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cr= classification_report(y_test,y_pred )
print(cr)
```

```python
print("Train-test split predictions (y_pred):")
print(pd.Series(y_pred).value_counts())

print("Final test predictions:")
print(pd.Series(test_predictions).value_counts())
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
y_probs = model.predict_proba(X_test)[:,1]
roc_auc_score(y_test, y_probs)
```

```python

from sklearn.metrics import roc_curve, auc

y_probs = model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)
print(roc_auc)

plt.figure()
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

precision, recall, _ = precision_recall_curve(y_test, y_probs)
ap_score = average_precision_score(y_test, y_probs)

plt.plot(recall, precision, label=f'AP = {ap_score:.2f}')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid()
plt.show()
```
