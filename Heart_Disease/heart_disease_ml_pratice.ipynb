##  Data Preprocessing

```python
x = df.drop('HeartDisease', axis = 1)
y = df['HeartDisease'].map({
    'Yes':1,
    'No':0
})
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
scaler  = StandardScaler()
x[df_num] = scaler.fit_transform(x[df_num])
```

```python
x.head()
```

## Feature Scaling: Standardization of Numerical Variables

To ensure consistent scale across features and improve model convergence, we applied **standardization** to the numerical variables using `StandardScaler` from `sklearn.preprocessing`. This transformation rescales each feature to have:

- A mean of 0
- A standard deviation of 1

The following numerical features were standardized:

- BMI
- PhysicalHealth
- MentalHealth
- SleepTime

```python


df_cat.remove('HeartDisease')
encoder = OneHotEncoder(drop='first', handle_unknown = 'ignore', sparse_output=False )

encoded = encoder.fit_transform(x[df_cat])

x_df = pd.DataFrame( encoded, columns = encoder.get_feature_names_out(df_cat), index = x.index )

x_encoded = pd.concat( [x.drop(columns = df_cat),x_df ], axis = 1 )

x_encoded.head()
```

## Categorical Encoding: One-Hot Encoding

To prepare the dataset for machine learning models, all categorical features must be converted into numerical format. We applied **One-Hot Encoding** to the categorical variables using `sklearn.preprocessing.OneHotEncoder`.

Key configuration details:
- `drop='first'` was used to avoid multicollinearity by removing the first category in each feature.
- `handle_unknown='ignore'` ensures that the encoder can handle unseen categories during inference without raising errors.
- The encoded result was returned as a dense array (`sparse_output=False`) and converted into a DataFrame with proper column names using `get_feature_names_out`.

After encoding, the original categorical columns were dropped and replaced with their encoded counterparts. The resulting feature matrix is now fully numerical and ready for model training.


```python


X_train, X_test, y_train, y_test = train_test_split(
    x_encoded, y, 
    test_size=0.2, 
    random_state=42, 
    stratify=y  # maintains class balance in both splits
)
```

## Train-Test Split

To evaluate model performance on unseen data, we split the dataset into training and test sets using an 80/20 ratio. 

The `train_test_split()` function from `sklearn.model_selection` was used with the following configuration:
- `test_size=0.2` to reserve 20% for testing
- `random_state=42` for reproducibility
- `stratify=y` to preserve the original class distribution of the target variable (`HeartDisease`)

This ensures both the training and test sets are representative of the overall dataset.


```python
# Initialize Random Forest with class_weight='balanced'
rf_model = RandomForestClassifier(
    n_estimators=100,               # number of trees
    max_depth=None,                 # let trees grow fully
    class_weight='balanced',        # handle imbalance
    random_state=42,
    n_jobs=-1                       # use all cores for faster training
)

# Train the model
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```


- The model achieves **high precision and recall for the majority class (Class 0)**, resulting in a strong F1-score of 0.95.
- For the minority class (Class 1), the **recall is only 0.11**, indicating that the model fails to correctly identify most heart disease cases.
- The **precision for Class 1 is 0.30**, meaning 70% of the predicted positives are actually false positives.
- The **F1-score for Class 1 is 0.16**, highlighting poor performance on the minority class.

### Overall Metrics:

- **Accuracy**: 90.01%
- **Macro Avg F1-Score**: 0.56
- **Weighted Avg F1-Score**: 0.88

### Conclusion:

Despite a high overall accuracy, the model is heavily biased toward the majority class.
This is evident from the low recall and F1-score for Class 1. 
In critical use cases such as health risk detection, this performance is inadequate, as **failing to identify true positive cases is costly**.

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", Counter(y_train))
print("After SMOTE:", Counter(y_train_smote))
```

```python
rf_smote_model = RandomForestClassifier(
    n_estimators=100,
    class_weight=None,     # No need since SMOTE already balances the data
    random_state=42,
    n_jobs=-1
)

rf_smote_model.fit(X_train_smote, y_train_smote)
y_pred_rf_smote = rf_smote_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_rf_smote))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf_smote))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf_smote))
```

```python
# Initialize with class_weight='balanced' to handle target imbalance
model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

## Final Model Comparison

This section summarizes the performance of three classification models in predicting heart disease. Each model is evaluated on the same imbalanced test dataset.

| Metric                      | Logistic Regression<br>(class_weight='balanced') | Random Forest | Random Forest + SMOTE |
|-----------------------------|---------------------------------------------------|----------------|------------------------|
| **Accuracy**                | 74.89%                                            | **90.15%**     | 88.13%                 |
| **Precision (Class 1)**     | 0.22                                              | 0.30           | 0.27                   |
| **Recall (Class 1)**        | **0.77** ✅                                       | 0.11           | 0.22                   |
| **F1-Score (Class 1)**      | **0.35** ✅                                       | 0.16           | 0.24                   |
| **False Negatives**         | **1,233** ✅                                      | 4,857          | 4,276                  |
| **True Positives**          | 4,242                                             | 618            | 1,199                  |

###  Insights

- **Logistic Regression with class weights** had the **best recall and F1-score** for the minority class, making it the most reliable model for detecting heart disease.
- **Random Forest (vanilla)** had the **highest overall accuracy**, but its recall on the positive class was the **worst**, leading to many missed heart disease cases.
- **Random Forest with SMOTE** provided a **better balance** than vanilla RF, but still underperformed compared to Logistic Regression in identifying true positives.

---

### Conclusion

If the **priority is identifying heart disease cases (sensitivity/recall)** — as in most healthcare settings — then:

> **Logistic Regression (with class weights)** is the best choice despite lower overall accuracy.

If **overall accuracy** or minimizing false positives is the primary goal, Random Forest may be preferred.
